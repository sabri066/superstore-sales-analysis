#  Superstore Sales Analysis

###  Objective
Analyze sales and profit data to identify business trends and insights.

```python
import pandas as pd

df = pd.read_csv('Sample - Superstore.csv')

df.head()
```

## 🛠 Tools Used
- Python
- Pandas
- Matplotlib

```python
df.info()
```

```python
df.isnull().sum()
```

##  Process
- Loaded dataset
- Cleaned data
- Created new features (Month, Year)
- Performed analysis
- Built visualizations

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year
```

```python
df.groupby('Region')['Profit'].sum()
```

```python
df.groupby('Category')['Sales'].sum()
```

```python
df.groupby('Month')['Sales'].sum()
```

```python
import matplotlib.pyplot as plt
```

```python
df.groupby('Region')['Sales'].sum().plot(kind='bar')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
df.groupby('Category')['Profit'].sum().plot(kind='bar')
plt.title('Profit by Category')
plt.xticks(rotation=0)
plt.show()
```

```python
df.groupby('Month')['Sales'].sum().plot(kind='line')
plt.title('Monthly Sales Trend')
plt.show()
```

```python
print("Insights:")
print("1. The region with highest sales is Western Europe")
print("2. The most profitable category is Technology")
print("3. Sales peak during month 12")
```
